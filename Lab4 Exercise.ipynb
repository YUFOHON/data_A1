### Q1. Load the 'data.csv' into a pandas DataFrame called df. Show the distribution of each feature (i.e. column) using histograms.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
f, ax = plt.subplots(nrows=5,ncols=1,figsize=(10,10))
ax[0].hist(df["x1"])
ax[1].hist(df["x2"])
ax[2].hist(df["x3"])
ax[3].hist(df["x4"])
ax[4].hist(df["x5"])
```

```python
df.boxplot()

def outliers_iqr(ys):
    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = (ys <= lower_bound) | (ys >= upper_bound)
    
    if True in outliers:
     print("my_list contains outliers")
    else:
     print("my_list does not contain outliers")

outliers_iqr(df["x1"])
outliers_iqr(df["x2"])
outliers_iqr(df["x3"])
outliers_iqr(df["x4"])
outliers_iqr(df["x5"])
```

### Q3. Use probability plots to determine which features are from normal distribution and which features are from uniform distribution. Write down your answers based on the generated figures. 
#### 3.1 Features from normal distribution: x1, x2, x3, x5
#### 3.2 Features from uniform distribution:  x3

```python
import scipy.stats as stats
def normality_plot(ys):
    stats.probplot(ys, dist='norm', plot=plt);
    plt.xlabel('Theoretical Quantiles (Normal Distribution)');
    plt.show()
    plt.figure()
    stats.probplot(ys, dist='uniform', plot=plt);
    plt.xlabel('Theoretical Quantiles (Uniform Distribution)');
    plt.show()

normality_plot(df["x1"])
normality_plot(df["x2"])
normality_plot(df["x3"])
normality_plot(df["x4"])
normality_plot(df["x5"])

```
